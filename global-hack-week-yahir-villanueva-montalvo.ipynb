# Challenge #3 

This challenge will be a random number generator


```python
import random
 
# This is the order of the list before getting randomized
li = [34, 12, 80, 61, 15, 69]
 
#This code is showing what the list looked liked before being randomized
print("The list before being randomized is : ", end="")
for i in range(0, len(li)):
    print(li[i], end=" ")
print("\r")
 
# using shuffle(li) to randomize the list
random.shuffle(li)
 
# This code will randomize the numbers above
print("The list after being randomized is : ", end="")
for i in range(0, len(li)):
    print(li[i], end=" ")
print("\r")
 
# using uniform() to generate random floating number in range
# prints number between 1 and 100
print("The random floating point number between 1 and 100 is : ", end="")
print(random.uniform(1, 100))
```
